## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from sklearn.preprocessing import OneHotEncoder

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
display(attrition_df.head())
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df=attrition_df[["Attrition","Department"]]
y_df
```

```python
y_df["Department"].value_counts()
```

```python
# Create a list of at least 10 column names to use as X data
X_columns=['Age', 'DistanceFromHome',
       'Education', 'EnvironmentSatisfaction', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'NumCompaniesWorked', 'PercentSalaryHike',
       'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion' ]


# Create X_df using your selected columns

X_df= attrition_df[X_columns]
X_df

# Show the data types for X_df
X_df.dtypes
```

```python
attrition_df.columns
```

```python
from re import X
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Create a StandardScaler
scaler = StandardScaler()



# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit_transform(X_train)

# Scale the training and testing data
X_test_scaled = scaler.transform(X_test)
X_train_scaled =scaler.transform(X_train)
```

```python
y_test['Department'].shape
```

```python
# Create a OneHotEncoder for the Department column
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
encoder.fit(np.array(y_train["Department"]).reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_department = encoder.transform(np.array(y_train["Department"]).reshape(-1, 1))
y_test_encoded_department = encoder.transform(np.array(y_test["Department"]).reshape(-1, 1))
y_train_encoded_department


```

```python
# Create a OneHotEncoder for the Attrition column
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit the encoder to the training data
encoder.fit(np.array(y_train["Attrition"]).reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_attrition = encoder.transform(np.array(y_train["Attrition"]).reshape(-1, 1))
y_test_encoded_attrition = encoder.transform(np.array(y_test["Attrition"]).reshape(-1, 1))
y_train_encoded_attrition
```

## Create, Compile, and Train the Model

```python
len(X_train.columns)
```

```python
# Find the number of columns in the X training data
len(X_train.columns)

# Create the input layer
input_nodes = len(X_train.columns)
input_layer = layers.Input(shape=(input_nodes,))

# Create at least two shared layers
shared_layer_1 = layers.Dense(units=128, activation='relu',name="shared_layer_1")(input_layer)
shared_layer_2 = layers.Dense(units=64, activation='relu', name='shared_layer_2')(shared_layer_1)

```

```python
# Create a branch for Department
# with a hidden layer and an output layer
hidden_layer_1 = layers.Dense(units=64, activation='relu',name='hidden_layer_1')(shared_layer_2)

# Create the hidden layer


# Create the output layer

output_layer_Department = layers.Dense(units=3, activation='softmax',name='output_layer_Department')(hidden_layer_1)
```

```python
units=len(y_train_encoded_department[0])
print(units)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
hidden_layer_2 = layers.Dense(units=64, activation='relu',name='hidden_layer_2')(shared_layer_2)
# Create the hidden layer


# Create the output layer

output_layer_Attrition = layers.Dense(units=len(y_train_encoded_attrition[0]), activation='sigmoid',name='output_layer_Attrition')(hidden_layer_2)
```

```python
print(len(y_train_encoded_attrition[0]))
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[output_layer_Department, output_layer_Attrition])

# Compile the model
model.compile(optimizer='adam',
              loss={"output_layer_Department":'categorical_crossentropy',
                    "output_layer_Attrition":'binary_crossentropy'},
              metrics={"output_layer_Department":'accuracy',
                       "output_layer_Attrition":'accuracy'})


# Summarize the model
model.summary()
```

```python
# Reshape your target arrays to have shape (num_samples, num_classes)
#y_train_encoded_department = y_train_encoded_department.reshape(1102, -1)
#y_train_encoded_attrition = y_train_encoded_attrition.reshape(1102, -1)

# Reshape your test target arrays as well to match the training data
#y_test_encoded_department = y_test_encoded_department.reshape(y_test_encoded_department.shape[1], -1)
#y_test_encoded_attrition = y_test_encoded_attrition.reshape(y_test_encoded_attrition.shape[1], -1)

# Now fit the model
history=model.fit(X_train_scaled, {"output_layer_Department": y_train_encoded_department,
                                   "output_layer_Attrition": y_train_encoded_attrition},
                  epochs=100, batch_size=64)
                 # validation_data=(X_test_scaled, [y_test_encoded_department, y_test_encoded_department]))
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {"output_layer_Department": y_test_encoded_department,
                                   "output_layer_Attrition": y_test_encoded_attrition})
```
